## Vamos a intentar hacer algo con los modelos, en primera instancia vamos a probar un problema de clasificacion

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
vueltas = pd.read_csv('../datasets_completos/dataset_spain_ham.csv')
vueltas
```

```python
from datetime import datetime
```

```python
vueltas = vueltas[vueltas['LapTime'] != '0']
```

```python
vueltas
```

```python
vueltas['LapTime'] = pd.to_datetime(vueltas['LapTime'], format='%M:%S.%f')
```

```python
vueltas
```

```python
def convert_to_seconds(t):
    return (t - datetime(1900,1,1)).total_seconds()
```

```python
vueltas
```

```python
vueltas['Laptime'] = vueltas['LapTime'].apply(convert_to_seconds)
```

```python
import math
```

```python
def tiempo_a_float(tiempo):
    try:
        print(tiempo)
        minutos, segundos = map(float, tiempo.split(':'))
        segundos = str(segundos)
        segundos, milisegundos = map(float, segundos.split('.'))
        segundos_totales = minutos * 60 + segundos + milisegundos / 1000
        return segundos_totales
    except:
        print("Era nan")
```

```python
vueltas['tiempoPilotoDelante'] = vueltas['tiempoPilotoDelante'].apply(tiempo_a_float)
vueltas['tiempoPilotoDetras'] = vueltas['tiempoPilotoDetras'].apply(tiempo_a_float)
```

```python
vueltas
```

```python
vueltas.drop(['LapTime'], axis=1, inplace=True)
```

```python
vueltas
```

```python
# Creo variables dummies (one hot encoding)

dummies_compound = pd.get_dummies(vueltas['Compound'], prefix='compound')
vueltas = pd.concat([vueltas, dummies_compound], axis=1)
vueltas.drop('Compound', axis=1, inplace=True)
vueltas['FreshTyre'] = vueltas['FreshTyre'].astype(int)
vueltas['Rainfall'] = vueltas['Rainfall'].astype(int)

#Vamos a hacer una aproximacion sin utilizar lo de pilotoDelante y pilotoDetras, solo el tiempo
vueltas.drop(['pilotoDelante', 'pilotoDetras' ], axis=1, inplace=True)


```

```python
vueltas
```

```python
vueltas = vueltas.dropna()
vueltasFinal = vueltas
```

```python
vueltas
```

### Ahora hemos de preparar los dataset, con los datos de entrenamiento y los datos de prueba

```python
X_train, X_test, y_train, y_test = train_test_split(vueltas.drop('makeStop', axis=1), 
                                                    vueltas['makeStop'], 
                                                    test_size=0.2, 
                                                    random_state=42)
```

```python
# entrenar modelo de árboles de clasificación
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# hacer predicciones con datos de prueba
y_pred = clf.predict(X_test)
```

```python
X_test
```

```python
X_test['makeStopPred'] = y_pred
```

```python
X_test.to_csv("prediccion1.csv")
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score

# evaluar desempeño del modelo
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('F1-score:', f1_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
vueltas_normalizadas = scaler.fit_transform(vueltasFinal)
```

```python
train_size = int(len(vueltas_normalizadas) * 0.8)
test_size = len(vueltas_normalizadas) - train_size
train_data, test_data = vueltas_normalizadas[0:train_size,:], vueltas_normalizadas[train_size:len(vueltas_normalizadas),:]
```

```python
def create_sequences(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), :]
        X.append(a)
        Y.append(dataset[i + look_back, :])
    return np.array(X), np.array(Y)
```

```python
# Crear secuencias para LSTM
look_back = 10 # número de filas anteriores a considerar
train_X, train_Y = create_sequences(train_data, look_back)
test_X, test_Y = create_sequences(test_data, look_back)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
# Definir modelo LSTM
model = Sequential()
model.add(LSTM(50, input_shape=(look_back, 19)))
model.add(Dense(19))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(train_X, train_Y, epochs=100, batch_size=64, verbose=2)
```

```python
# Predicción en conjunto de prueba
test_predict = model.predict(test_X)

# Desnormalización de datos
test_predict = scaler.inverse_transform(test_predict)
test_Y = scaler.inverse_transform(test_Y)

# Cálculo del error de predicción
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test_Y, test_predict)
print('MSE:', mse)
```
