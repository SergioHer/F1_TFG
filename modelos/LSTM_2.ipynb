```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Cargar los datos en un DataFrame de pandas
df = pd.read_csv('./vueltas_ham_final.csv')

# Dividir los datos en conjuntos de entrenamiento y prueba
train_data = df[df['anyo'] <= 2021]
test_data = df[df['anyo'] == 2022]
```

```python
train_data
```

```python
from sklearn.preprocessing import StandardScaler

# Normalizar las caracterÃ­sticas
scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)
```

```python
data = pd.DataFrame(train_data)
data.iloc[1][6]
```

```python
import numpy as np
import tensorflow as tf
# Crear una secuencia de 10 vueltas como entrada y la siguiente vuelta como salida
def create_sequences(data):
    X, y = [], []
    for i in range(len(data)-11):
        X.append(data[i:i+10])
        y.append(data[i+10, 6])
    return np.array(X), np.array(y)


```

```python
X_train, y_train = create_sequences(train_data)
X_test, y_test = create_sequences(test_data)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Construir el modelo
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(10, 19)))
model.add(Dense(2, activation='sigmoid'))

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
X_train
```

```python
# Entrenar el modelo
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
score = model.evaluate(X_test, y_test, batch_size=32)
print('Test accuracy:', score[1])
```
